```python
import numpy as np

#Datos de ventas mensuales
ventas_por_mes = np.array([750, 800, 830, 810, 860, 930, 990, 1040, 1090, 1140, 1190, 1240])

# Calcular la media, la mediana, la desviación estándar y la varianza de las ventas mensuales
media_ventas, mediana_ventas, std_ventas, varianza_ventas = ventas_por_mes.mean(), np.median(ventas_por_mes), ventas_por_mes.std(), ventas_por_mes.var()

# Mostrar los resultados
print("La media de ventas mensuales es:", media_ventas)
print("La mediana de ventas mensuales es:", mediana_ventas)
print("La desviación estándar de ventas mensuales es:", std_ventas)
print("La varianza de ventas mensuales es:", varianza_ventas)
```

```python
import pandas as pd

# Cargar los datos de Excel
df = pd.read_excel("/content/drive/MyDrive/ANALITICA DATOS/Análisis de Ventas Mensuales en una Licorera.xlsx")

# Definir la columna 'Artículo' como el índice
df = df.set_index('Articulo')

# Determinar la cantidad total de ventas por artículo
ventas_articulo = df.sum(axis=1)

# Encontrar los artículos con las ventas más altas
articulos_max_ventas = ventas_articulo[ventas_articulo == ventas_articulo.max()].index.tolist()

# Presentar los artículos con las ventas más destacadas
print("El artículo con la venta más alta es:", articulos_max_ventas)
```

```python
import pandas as pd

# Cargar los datos de Excel
df = pd.read_excel("/content/drive/MyDrive/ANALITICA DATOS/Análisis de Ventas Mensuales en una Licorera.xlsx")

# Determinar la totalidad de ventas por mes
ventas_por_mes = df.sum()[1:]

# Transformar la serie a tipo numérico y reemplazar los valores faltantes con ceros
ventas_por_mes_numerico = pd.to_numeric(ventas_por_mes, errors='coerce').fillna(0)

# Encontrar el mes con las ventas más altas
mes_max_ventas = ventas_por_mes_numerico.idxmax()

# Mostrar el mes con las ventas más altas
print("El mes con la venta más alta es:", mes_max_ventas)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Establecer el estilo de Matplotlib
sns.set_style("whitegrid")

# Cargar los datos de Excel
df = pd.read_excel("/content/drive/MyDrive/ANALITICA DATOS/Análisis de Ventas Mensuales en una Licorera.xlsx", index_col='Articulo')

# Transponer el DataFrame y graficar
plt.figure(figsize=(12, 8))
df.T.plot(ax=plt.gca(), marker='o')

# Añadir etiquetas y título
plt.xlabel('Mes')
plt.ylabel('Cantidad de ventas')
plt.title('Análisis de ventas mensuales')
plt.xticks(rotation=45)
plt.legend(title='Artículo')
plt.tight_layout()

# Mostrar el gráfico
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos de Excel
df = pd.read_excel("/content/drive/MyDrive/ANALITICA DATOS/Análisis de Ventas Mensuales en una Licorera.xlsx")

#Determinar la totalidad de las ventas mensuales y generar el gráfico de barras
plt.figure(figsize=(10, 6))
df.sum()[1:].plot(kind='bar', color='purple')

# Añadir etiquetas y título
plt.xlabel('Mes')
plt.ylabel('Total de ventas')
plt.title('Análisis de ventas totales por mes')

# Indicar el gráfico
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
